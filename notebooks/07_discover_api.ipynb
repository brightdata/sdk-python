# 🔎 Discover API - AI-Powered Web Search

Test the Discover API — web search with AI-powered relevance ranking:
- Basic search
- Intent-based relevance ranking
- Filtering with keywords, country, city
- Full-page content extraction
- Manual trigger/poll/fetch control

---

## Setup

```python
import os
from dotenv import load_dotenv
load_dotenv()

API_TOKEN = os.getenv("BRIGHTDATA_API_TOKEN")
if not API_TOKEN:
    raise ValueError("Set BRIGHTDATA_API_TOKEN in .env file")

print(f"API Token: {API_TOKEN[:10]}...{API_TOKEN[-4:]}")
print("Setup complete!")
```

## Initialize Client

```python
from brightdata import BrightDataClient

client = BrightDataClient(token=API_TOKEN)

print("Client initialized")
```

---
## Test 1: Basic Discover Search

Simple search query — returns AI-ranked results with relevance scores.

```python
QUERY = "artificial intelligence trends 2026"

print(f"Discover Search: '{QUERY}'\n")

async with client:
    result = await client.discover(
        query=QUERY
    )

print(f"Success: {result.success}")
print(f"Total Results: {result.total_results}")
print(f"Task ID: {result.task_id}")
print(f"Duration: {result.duration_seconds}s" if result.duration_seconds else "Duration: N/A")

if result.success and result.data:
    print("\n--- Top 10 Results ---")
    for i, item in enumerate(result.data[:10]):
        score = item.get('relevance_score', 0)
        title = item.get('title', 'N/A')
        link = item.get('link', 'N/A')
        desc = item.get('description', 'N/A')
        print(f"\n{i+1}. [{score:.2f}] {title}")
        print(f"   URL: {link}")
        print(f"   {desc[:80]}..." if len(str(desc)) > 80 else f"   {desc}")
else:
    print(f"\nError: {result.error}")
```

---
## Test 2: Search with Intent

The `intent` parameter separates *what* you're searching from *why*.
The API uses this to rank results by semantic relevance, not just keyword matching.

```python
QUERY = "Tesla battery technology"
INTENT = "recent breakthroughs in EV battery chemistry"

print(f"Query: '{QUERY}'")
print(f"Intent: '{INTENT}'\n")

async with client:
    result = await client.discover(
        query=QUERY,
        intent=INTENT
    )

print(f"Success: {result.success}")
print(f"Total Results: {result.total_results}")

if result.success and result.data:
    print("\n--- Results ranked by relevance to intent ---")
    for i, item in enumerate(result.data[:10]):
        score = item.get('relevance_score', 0)
        title = item.get('title', 'N/A')
        link = item.get('link', 'N/A')
        print(f"\n{i+1}. [{score:.4f}] {title}")
        print(f"   {link}")
else:
    print(f"\nError: {result.error}")
```

---
## Test 3: Filter Keywords & Localization

Use `filter_keywords` to narrow results, and `country`/`city` for localization.

```python
QUERY = "sustainable fashion brands"
INTENT = "eco-friendly clothing companies"
FILTER_KEYWORDS = ["sustainability", "eco-friendly", "organic"]

print(f"Query: '{QUERY}'")
print(f"Intent: '{INTENT}'")
print(f"Filter Keywords: {FILTER_KEYWORDS}")
print(f"Country: us\n")

async with client:
    result = await client.discover(
        query=QUERY,
        intent=INTENT,
        filter_keywords=FILTER_KEYWORDS,
        country="us",
        num_results=10
    )

print(f"Success: {result.success}")
print(f"Total Results: {result.total_results}")

if result.success and result.data:
    print("\n--- Filtered Results ---")
    for i, item in enumerate(result.data[:10]):
        score = item.get('relevance_score', 0)
        title = item.get('title', 'N/A')
        link = item.get('link', 'N/A')
        print(f"\n{i+1}. [{score:.2f}] {title}")
        print(f"   {link}")
else:
    print(f"\nError: {result.error}")
```

---
## Test 4: Include Page Content

Set `include_content=True` to get full page content as markdown alongside each result.

```python
QUERY = "python asyncio tutorial"
INTENT = "beginner-friendly guide to async programming in Python"

print(f"Query: '{QUERY}'")
print(f"Intent: '{INTENT}'")
print("Include Content: True\n")

async with client:
    result = await client.discover(
        query=QUERY,
        intent=INTENT,
        include_content=True,
        num_results=3
    )

print(f"Success: {result.success}")
print(f"Total Results: {result.total_results}")

if result.success and result.data:
    for i, item in enumerate(result.data[:3]):
        score = item.get('relevance_score', 0)
        title = item.get('title', 'N/A')
        content = item.get('content')
        print(f"\n{'='*60}")
        print(f"{i+1}. [{score:.2f}] {title}")
        print(f"   URL: {item.get('link', 'N/A')}")
        if content:
            preview = content[:200].replace('\n', ' ')
            print(f"   Content preview: {preview}...")
        else:
            print("   Content: None")
else:
    print(f"\nError: {result.error}")
```

## Test 5: Manual Trigger / Poll / Fetch

Use `discover_trigger()` for manual control — trigger the search, do other work, then poll and fetch.

```python
QUERY = "market research SaaS pricing 2026"
INTENT = "competitor pricing strategies for B2B SaaS"

print(f"Query: '{QUERY}'")
print(f"Intent: '{INTENT}'\n")

async with client:
    # Step 1: Trigger the search
    job = await client.discover_trigger(
        query=QUERY,
        intent=INTENT,
        num_results=5
    )
    print(f"Step 1 - Triggered: {job}")
    print(f"   Task ID: {job.task_id}")

    # Step 2: Check status
    status = await job.status()
    print(f"\nStep 2 - Status: {status}")

    # Step 3: Wait for completion
    print("\nStep 3 - Waiting for results...")
    final_status = await job.wait(timeout=60, poll_interval=2)
    print(f"   Final status: {final_status}")

    # Step 4: Fetch results
    data = await job.fetch()
    print(f"\nStep 4 - Fetched {len(data)} results")

    for i, item in enumerate(data):
        score = item.get('relevance_score', 0)
        title = item.get('title', 'N/A')
        link = item.get('link', 'N/A')
        print(f"\n  {i+1}. [{score:.2f}] {title}")
        print(f"     {link}")
```

## Summary

### Methods

| Method | Description |
|--------|-------------|
| `client.discover(query, ...)` | Search + poll + return results |
| `client.discover_trigger(query, ...)` | Trigger search, return `DiscoverJob` for manual control |

### Parameters

| Parameter | Type | Description | Default |
|-----------|------|-------------|--------|
| `query` | str | Search query | Required |
| `intent` | str | Why you're searching — guides AI relevance ranking | `None` |
| `include_content` | bool | Return page content as markdown | `False` |
| `country` | str | Country code for localized results | `None` |
| `city` | str | City for localized results | `None` |
| `language` | str | Language code | `None` |
| `filter_keywords` | list[str] | Filter results by keywords | `None` |
| `num_results` | int | Number of results to return | `None` |
| `format` | str | Response format | `"json"` |
| `timeout` | int | Max seconds to wait | `60` |
| `poll_interval` | int | Seconds between status checks | `2` |

### DiscoverJob Methods (Manual Control)

| Method | Description |
|--------|-------------|
| `job.status()` | Check task status (`'processing'` or `'done'`) |
| `job.wait(timeout, poll_interval)` | Poll until done or timeout |
| `job.fetch()` | Fetch results after completion |
| `job.to_result()` | Wait + fetch + wrap as `DiscoverResult` |

### Result Item Fields

| Field | Type | Description |
|-------|------|-------------|
| `link` | str | URL of the result |
| `title` | str | Page title |
| `description` | str | Snippet / meta description |
| `relevance_score` | float | 0.0–1.0, AI-computed relevance to stated intent |
| `content` | str/null | Page content in markdown (only if `include_content=True`) |

### Discover vs SERP

| | SERP (`client.search.google`) | Discover (`client.discover`) |
|---|---|---|
| Results | Raw search engine results | AI-ranked by intent |
| Relevance | Keyword-based | Semantic (via `intent`) |
| Content | Snippets only | Optional full-page markdown |
| Use case | Known search queries | Open-ended research |
