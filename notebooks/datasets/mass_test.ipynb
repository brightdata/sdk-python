# Datasets Mass Test

Tests all 87 datasets (excluding Amazon, LinkedIn, Crunchbase, Instagram, TikTok, YouTube).
Each dataset has 2 cells:
1. **Trigger** - Creates a snapshot using `sample()` and returns snapshot_id
2. **Download** - Fetches data when snapshot is ready

Each dataset uses unique variable names so cells can be run in any order.

```python
from brightdata import BrightDataClient

client = BrightDataClient()
await client.__aenter__()
```
